```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn.linear_model import LinearRegression

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on='Mouse ID',how='right')
#combined_data = combined_data[['Mouse ID', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Drug Regimen', 'Sex', 'Age_months', 'Weight (g)']]

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
num_mice = combined_data['Mouse ID'].nunique()
print(f"Number of mice: {num_mice}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# duplicate_mice = combined_data[combined_data.duplicated(['Mouse ID', 'Timepoint'])]
# print("Duplicate Mice:")
# print(duplicate_mice)
#array(['g989'], dtype=object)
duplicate_mice = combined_data[combined_data.duplicated(['Mouse ID', 'Timepoint'], keep=False)]
duplicate_mouse_ids = duplicate_mice['Mouse ID'].unique()

print(f"Duplicate Mouse IDs: {duplicate_mouse_ids} dtype={duplicate_mouse_ids.dtype}")
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = combined_data[combined_data['Mouse ID'].isin(duplicate_mice['Mouse ID'])]
print("Duplicate Mouse Data:")
print(duplicate_mouse_data)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_data = combined_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
print(clean_combined_data.head())
```

```python
# Checking the number of mice in the clean DataFrame.
num_clean_mice = clean_combined_data['Mouse ID'].nunique()
print(f"Number of mice in clean DataFrame: {num_clean_mice}")
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_combined_data.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(last_timepoint, clean_combined_data, on=['Mouse ID', 'Timepoint'])
```

```python
# Put treatments into a list for for loop (and later for plot labels)


# Create empty list to fill with tumor vol data (for plotting)


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes

    
    # add subset 

    
    # Determine outliers using upper and lower bounds
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

tumor_vol_data = []

for treatment in treatments:
    subset = final_tumor_volume[final_tumor_volume['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    quartiles = subset.quantile([0.25, 0.5, 0.75])
    lower_bound = quartiles[0.25] - 1.5 * (quartiles[0.75] - quartiles[0.25])
    upper_bound = quartiles[0.75] + 1.5 * (quartiles[0.75] - quartiles[0.25])
    outliers = subset[(subset < lower_bound) | (subset > upper_bound)]
    tumor_vol_data.append(subset)
    print(f"{treatment} Potential Outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_data, labels=treatments)
plt.title('Distribution of Tumor Volume for Each Treatment Group')
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_data = clean_combined_data[clean_combined_data['Drug Regimen'] == 'Capomulin']
single_mouse_data = capomulin_data[capomulin_data['Mouse ID'] == 's185']
plt.plot(single_mouse_data['Timepoint'], single_mouse_data['Tumor Volume (mm3)'], marker='o')
plt.title('Capomulin treatment of mouse 1509')
plt.xlabel('Time Point(days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = mouse_weight.corr(avg_tumor_volume)
model = LinearRegression().fit(mouse_weight.values.reshape(-1, 1), avg_tumor_volume.values.reshape(-1, 1))
slope = model.coef_[0][0]
intercept = model.intercept_
print(f"Correlation Coefficient: {correlation}")
print(f"Linear Regression Model: y = {slope}x + {intercept}")

# Generate a scatter plot with linear regression line
plt.scatter(mouse_weight, avg_tumor_volume)
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Plot the linear regression line
plt.plot(mouse_weight, model.predict(mouse_weight.values.reshape(-1, 1)), color='red', linewidth=2)

# Show the plot
plt.show()
```
